```python
import matplotlib.pyplot as plt
import numpy as np

# Plot histograms for all features
def plot_histograms(df, bin_width=10):
    num_features = df.select_dtypes(include=['number']).columns  # Select numeric columns
    num_cols = 3  # Number of columns in the subplot grid
    num_rows = int(np.ceil(len(num_features) / num_cols))  # Calculate the number of rows

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
    axes = axes.flatten()  # Flatten the axes for easy iteration

    for i, col in enumerate(num_features):
        ax = axes[i]
        
        # Determine the bin size dynamically based on the range of the data
        col_range = df[col].max() - df[col].min()
        bins = int(np.ceil(col_range / bin_width))
        
        ax.hist(df[col].dropna(), bins=bins, color='blue', alpha=0.7, edgecolor='black')
        ax.set_title(f'Histogram of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# Call the function with your DataFrame
plot_histograms(data, bin_width=10)
```
